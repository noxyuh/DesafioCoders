```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('VendasdeCursos.csv')
```

```python
df = pd.DataFrame(dados)
```

```python
df
```

```python
df ['Receita Total'] = df ['Quantidade de Vendas'] * df['Preço Unitário']
```

```python
df
```

```python
top_curso = df.nlargest(1, 'Quantidade de Vendas').iloc[0]

print(f"O curso com o maior vendas é: {top_curso['Nome do Curso']} ")
```

```python
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df = df.sort_values(by='Data')
plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['Quantidade de Vendas'], marker='o', linestyle='-')
plt.title('Distribuição das Vendas ')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
